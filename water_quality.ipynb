```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("data.csv")
data.head(10)
```

```python
#data.drop(data[data["Mooring_Id"]!="A"].index)[one way of deleting row with specific value]
data=data[data["Mooring_Id"]=="A"]#second way but efficient
data=data[data.Water_Depth==7]
data=data[data.Temperature!=-9999]
data=data[data.Salinity!=-9999]
data=data[data.pH!=-9999]
```

```python
data.columns
```

```python
l=["Mooring_Id","Site_Description","Latitude","Longitude","Start_Date","End_Date","Deployment_No","Water_Depth","Sensor_ID","Date","Time"]
data=data.drop(l,axis=1)
```

```python
data.describe(include="all")
```

```python
data["Temperature"]=data["Temperature"]+15
y_data=data["Oxygen"]
data=data.drop("Oxygen",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_data=pd.DataFrame(scaler.fit_transform(data))
```

```python
x_data
```

```python
scaler1 = StandardScaler()
y_data=(scaler1.fit_transform(y_data.values.reshape(-1,1)))
```

```python
y_data[0]
```

```python
data.info()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
```

```python
plt.plot(x_data[0])
plt.title("Temperature")
```

```python
plt.plot(x_data[1])
plt.title("Salinity")
```

```python
fig,ax=plt.subplots(1,2,figsize=(13,6))
ax[0].plot(x_data[2])
ax[0].title.set_text("pH")
ax[1].plot(y_data)#pH and oxygen shows high degree of correlation
ax[1].title.set_text("Dissolved Oxygen")
```

```python
from sklearn.svm import SVR
model=SVR(C=1,kernel="linear")
model.fit(x_train,y_train)
print(model.score(x_train,y_train))
#accuracy 91.5%
```

```python
#too bad for this data
import lightgbm as lgb
d_train = lgb.Dataset(x_train, label=y_train)
params = {}
params['learning_rate'] = 0.003
params['boosting_type'] = 'gbdt'
params['objective'] = 'regression'
params['metric'] = 'mse'
params['sub_feature'] = 0.5
params['num_leaves'] = 100
params['min_data'] = 50
params['max_depth'] = 100
model = lgb.train(params,d_train,500)
```

```python
from sklearn import neighbors
model = neighbors.KNeighborsRegressor(n_neighbors =8,n_jobs=-1)
model.fit(x_train, y_train)
print(model.score(x_train,y_train))#best model
```

```python
y_predict = model.predict(x_test)
y_predict
```
